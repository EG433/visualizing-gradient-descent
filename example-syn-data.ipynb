```python
import numpy as np
import json

def random_data(samples, classes):
    X = np.zeros((samples * classes, 2))
    y = np.zeros(samples * classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(samples * class_number, samples * (class_number + 1))
        X[ix] = np.c_[np.random.rand(samples), np.random.rand(samples)]
        y[ix] = class_number
    return X, y

def vertical_data(samples, classes):
    X = np.zeros((samples*classes, 2))
    y = np.zeros(samples*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(samples*class_number, samples*(class_number+1))
        X[ix] = np.c_[np.random.randn(
            samples)*.1 + (class_number)/3, np.random.randn(samples)*.1 + 0.5]
        y[ix] = class_number
    return X, y

def spiral_data(samples, classes):
    X = np.zeros((samples*classes, 2))
    y = np.zeros(samples*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(samples*class_number, samples*(class_number+1))
        r = np.linspace(0.0, 1, samples)  # radius
        t = np.linspace(class_number*4, (class_number+1)*4,
                        samples) + np.random.randn(samples)*0.2
        X[ix] = np.c_[r*np.sin(t*2.5), r*np.cos(t*2.5)]
        y[ix] = class_number
    return X, y
```

```python
samples = 50
classes = 2

X_random, y_random = random_data(samples, classes)
random_dataset = [{'id': i + 1, 'x': float(X_random[i, 0]), 'y': float(X_random[i, 1]), 'class': int(y_random[i])} for i in range(len(X_random))]

X_spiral, y_spiral = spiral_data(samples, classes)
spiral_dataset = [{'id': i + 1, 'x': float(X_spiral[i, 0]), 'y': float(X_spiral[i, 1]), 'class': int(y_spiral[i])} for i in range(len(X_spiral))]

X_vertical, y_vertical = vertical_data(samples, classes)
vertical_dataset = [{'id': i + 1, 'x': float(X_vertical[i, 0]), 'y': float(X_vertical[i, 1]), 'class': int(y_vertical[i])} for i in range(len(X_vertical))]

with open('./src/assets/random_data.json', 'w') as f:
    json.dump(random_dataset, f, indent=4)

with open('./src/assets/vertical_data.json', 'w') as f:
    json.dump(vertical_dataset, f, indent=4)

with open('./src/assets/spiral_data.json', 'w') as f:
    json.dump(spiral_dataset, f, indent=4)
```
